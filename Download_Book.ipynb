```
import requests
import urllib
import bs4 
import wget
import os

# Try to find equal names for your target
search_terms = ["kätzchen","Katzen","Katze","Mietze", "mietzekatze", "cat", "kitty"]
dir_path = "Images_"+search_terms[0]
i = 0 # image index
try:
  os.makedirs(dir_path, exist_ok=True)
  print("Directory created..")
except:
  pass

for term in search_terms:

  url = "https://www.google.de/search?q="+term+"&tbm=isch"
  
  res = requests.get(url)
  html = res.text
  res_code = res.status_code
  img_list = []
  f = None
  


  if (res_code == 200):
    soup = bs4.BeautifulSoup(html, "html.parser")

    for el in soup.find_all('img'):
      try:
        try:
          pic_url = el['src']
          
          name = dir_path + "/"+ term +"_"+str(i).zfill(3)+".jpg"
          i+=1
          #requests.get(pic_url, stream=True)
          #img_list.append(name)
          f = wget.download(pic_url, name)
          print(name,pic_url)        
        except:
          pass
        
      except:
        pass
  else:
    print("No Respones")
```
